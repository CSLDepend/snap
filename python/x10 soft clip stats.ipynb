```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import re
```

```python
data = open("d:/sequence/garvanx10/na12878_r1-chr1-10m.sam", "r")
nnext = 5
scliplen = []
sclipqual = []
sclipavgqual = []
sclipnextqual = []
sclipnextavgqual = []
keepqual = []
keepavgqual = []
sclipstart = re.compile(r'^([0-9]+)S.*')
sclipend = re.compile(r'.*[A-Z]([0-9]+)S$')
nlines = 0
nclip = 0
while True:
    line = data.readline()
    if line == "":
        break
    fields = [s.strip() for s in line.split('\t')]
    cigar = fields[5]
    qual = fields[10]
    keep = qual
    m = sclipstart.match(cigar)
    if m != None:
        n = int(m.groups()[0])
        #print ("start", n, cigar, qual)
        scliplen.append(n)
        q = [ord(c)-33 for c in qual[:n]]
        sclipqual.extend(q)
        sclipavgqual.append(np.mean(q))
        keep = keep[n:]
        q = [ord(c)-33 for c in keep[:nnext]]
        sclipnextqual.extend(q)
        sclipnextavgqual.append(np.mean(q))
    m = sclipend.match(cigar)
    if m != None:
        n = int(m.groups()[0])
        #print ("end", n, cigar, qual)
        scliplen.append(n)
        q = [ord(c)-33 for c in qual[-n:]]
        sclipqual.extend(q)
        sclipavgqual.append(np.mean(q))
        keep = keep[:-n]
        q = [ord(c)-33 for c in keep[-nnext:]]
        sclipnextqual.extend(q)
        sclipnextavgqual.append(np.mean(q))
    if keep != qual:
        q = [ord(c)-33 for c in keep]
        keepqual.extend(q)
        keepavgqual.append(np.mean(q))
        nclip = nclip + 1
    nlines = nlines + 1
    # if nlines > 1000: break
"soft clipped %d of %d lines" % (nclip, nlines)
```

```python
plt.hist(scliplen)
plt.show()
```
